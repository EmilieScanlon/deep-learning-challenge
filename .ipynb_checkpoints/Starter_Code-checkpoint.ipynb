## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID column, 'EIN'
application_df = application_df.drop('EIN', axis=1)
application_df
```

```python
# Drop the non-beneficial ID column'NAME'.
application_df = application_df.drop('NAME', axis=1)
application_df
```

```python
# Determine the number of unique values in each column.
unique_counts = application_df.nunique()
print(unique_counts)
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
print(application_type_counts)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = application_df['APPLICATION_TYPE'].value_counts()[application_df['APPLICATION_TYPE'].value_counts() < 250].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_counts = application_df['CLASSIFICATION'].value_counts()
print(classification_counts)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts = application_df['CLASSIFICATION'].value_counts()
print(classification_counts[classification_counts > 1])
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
# Choose a cutoff value
cutoff = 1000

# Create a list of classifications to be replaced
classifications_to_replace = application_df['CLASSIFICATION'].value_counts()[application_df['CLASSIFICATION'].value_counts() < cutoff].index.tolist()

# Replace rare classifications with "Other"
application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].apply(lambda x: "Other" if x in classifications_to_replace else x)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df, columns=['CLASSIFICATION'])
```

```python
application_df.info()
```

```python
#Convert INCOME_AMT values into floats and display those values in application_df 
# 0-> 0
# 1-9999 ->1
# 10000-24999 -> 2
# 1M-5M -> 3

income_map = {
    '0': 0.0,
    '1-9999': 1.0,
    '10000-24999': 2.0,
    '100000-499999':3.0,
    '1M-5M': 4.0,
    '10M-50M': 5.0
}
application_df['INCOME_AMT'] = application_df['INCOME_AMT'].map(income_map)
application_df
```

```python
#Convert affiliation values into floats and display those values in application_df 
# Independent-> 0
# CompanySponsored	 ->1

AFFILIATION_map = {
    'Independent': 0.0,
    'CompanySponsored': 1.0
}
application_df['AFFILIATION'] = application_df['AFFILIATION'].map(AFFILIATION_map)
application_df
```

```python
#Convert CLASSIFICATION values into floats and display those values in application_df 

application_df['CLASSIFICATION_C1000'] = application_df['CLASSIFICATION_C1000'].astype('float')
application_df['CLASSIFICATION_C1200'] = application_df['CLASSIFICATION_C1200'].astype('float')
application_df['CLASSIFICATION_C2000'] = application_df['CLASSIFICATION_C2000'].astype('float')
application_df['CLASSIFICATION_C2100'] = application_df['CLASSIFICATION_C2100'].astype('float')
application_df['CLASSIFICATION_C3000'] = application_df['CLASSIFICATION_C3000'].astype('float')
application_df['CLASSIFICATION_Other'] = application_df['CLASSIFICATION_C3000'].astype('float')
application_df
```

```python
#Convert organization values into floats and display those values in application_df 
# Independent-> 0
# CompanySponsored	 ->1

ORGANIZATION_map = {
    'Association': 0.0,
    'Co-operative': 1.0,
    'Trust': 2.0
}
application_df['ORGANIZATION'] = application_df['ORGANIZATION'].map(ORGANIZATION_map)
application_df
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df[['STATUS', 'ASK_AMT', 'INCOME_AMT', 'AFFILIATION', 'CLASSIFICATION_C1000', 
                  'CLASSIFICATION_C1200', 'CLASSIFICATION_C2000', 'CLASSIFICATION_C2100', 
                    'CLASSIFICATION_C3000', 'CLASSIFICATION_Other', 'ORGANIZATION']]
y = application_df['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(80, activation='relu', input_shape=(11,)))

# Second hidden layer
nn.add(tf.keras.layers.Dense(30, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
nn.fit(X_train, y_train, epochs=10, batch_size=128, validation_data=(X_test, y_test))
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save('model.h5')
```
